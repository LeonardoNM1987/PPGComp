```python
import matplotlib.pyplot as plt

# Supondo que voc leu esses dados do seu arquivo txt
# Estes so apenas dados de exemplo
total_prepends = 100  # Valor hipottico
origem_prepends = [10, 20, 5, 40]  # Lista hipottica ao longo do tempo
intermed_prepends = [5, 1, 2, 5]  # Lista hipottica ao longo do tempo
unique_ases = [1, 3, 2, 4]  # Lista hipottica ao longo do tempo

# Datas correspondentes aos seus dados
dates = ['2010', '2013', '2016', '2020']  # As datas devem corresponder aos seus dados

# Criando o grfico
plt.plot(dates, origem_prepends, marker='o', label='Prepends Totais')
plt.plot(dates, intermed_prepends, marker='s', label='Prep.Origem')
plt.plot(dates, unique_ases, marker='x', label='Prep.Intermed.')

# Adicionando ttulo e rtulos
plt.title('Frao de ASes utilizando ASPP')
plt.xlabel('Perodo (Mensal, Jan2010-Abr2020)')
plt.ylabel('Frao de Ases Unicos')

# Adicionando legenda
plt.legend()

# Mostrando o grfico
plt.show()
```
